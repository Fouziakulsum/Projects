Data Preprocessing

```python
mod_df=daf.drop("Unnamed: 0",axis=1)
# Removes the column named 'Unnamed: 0' from the DataFrame
# axis=1 indicates that we are dropping a column (not a row)
mod_df.head()
```

Converting Date column into Day, month, year


```python
mod_df["Date"]=pd.to_datetime(mod_df["Date"])
# Converts the 'Date' column from string format to proper datetime format
# This helps in performing time-based operations (filtering, extracting year/month, etc.)
```

```python
day_loc=mod_df.columns.get_loc("Date")+1
# Finds the column index (position) of the 'Date' column
# Adds 1 so the new column will be inserted *right after* the 'Date' column
day_values=mod_df["Date"].dt.day
# Extracts the day (131) from each date in the 'Date' column
mod_df.insert(day_loc,"Day",day_values)
# Inserts the new 'Day' column into the DataFrame at the calculated position
mod_df
```

```python
month_loc=mod_df.columns.get_loc("Day")+1
# Finds the position of the 'Day' column
# Adds 1 so the new 'Month' column will be inserted right after 'Day'
month_values=mod_df["Date"].dt.month
# Extracts the month (112) from each date in the 'Date' column

mod_df.insert(month_loc,"Month",month_values)
# Inserts the new 'Month' column at the calculated position in the DataFrame

mod_df
```

```python
year_loc=mod_df.columns.get_loc("Month")+1
# Finds the index of the 'Month' column
# Adds 1 so the new 'Year' column is inserted right after 'Month'
year_values=mod_df["Date"].dt.year
# Extracts the year from each date in the 'Date' column

mod_df.insert(year_loc,"Year",year_values)
# Inserts the 'Year' column at the calculated position in the DataFrame

mod_df
```

```python
mod_df.drop("Date",axis=1,inplace=True)
mod_df
```

```python
mod_df.info()
```

```python
categorical_cols=mod_df.select_dtypes(include=["object"])
# Selects all columns from the DataFrame that have data type 'object'
# (usually string or categorical text data)
categorical_cols
```

```python
mod_df["Online Transactions Frequency"].unique()
# Returns all unique values present in the 'Online Transactions Frequency' column
# Useful for understanding the different categories or levels in this column
```

```python
mod_df["Merchant Reputation"].unique()
# Returns all unique values present in the 'Merchant Reputation' column
```

```python
frequency_map={"Low":0,"Medium":1,"High":2}
# Creates a mapping dictionary to convert 'Online Transactions Frequency' from text to numeric values

Reputation_map={"Bad":0,"Average":1,"Good":2}
# Creates a mapping dictionary to convert 'Merchant Reputation' from text to numeric values

mod_df["Online Transactions Frequency"]=mod_df["Online Transactions Frequency"].map(frequency_map)
# Replaces the text values in 'Online Transactions Frequency' with their corresponding numeric values

mod_df["Merchant Reputation"]=mod_df["Merchant Reputation"].map(Reputation_map)
# Replaces the text values in 'Merchant Reputation' with their corresponding numeric values

mod_df
```

```python
mod_df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
# Imports the OneHotEncoder class from scikit-learn
# Used to convert categorical variables into a one-hot (binary) encoded format
ohe=OneHotEncoder(sparse_output=False)
# Creates an instance of OneHotEncoder
# sparse_output=False ensures the output will be a dense NumPy array instead of a sparse matrix
```

```python
nominal_cols=mod_df.select_dtypes(include=["object"]).columns.tolist()
# Selects all columns with data type 'object' (categorical text columns)
# Converts them into a list for encoding
encoded=ohe.fit_transform(mod_df[nominal_cols])
# Applies OneHotEncoder to the nominal (categorical) columns
# fit_transform learns the categories and transforms them into a one-hot encoded array
encoded_df=pd.DataFrame(encoded,columns=ohe.get_feature_names_out(nominal_cols))
# Converts the one-hot encoded array into a DataFrame
# Uses get_feature_names_out to name the new columns based on original column names
encoded_mod_df=pd.concat([mod_df.drop(columns=nominal_cols),encoded_df],axis=1)
# Drops the original nominal columns from mod_df
# Concatenates the one-hot encoded columns to create a fully numeric DataFrame
encoded_mod_df.columns
```

```python
col=encoded_mod_df.pop("Is Fraudulent")
# Removes the 'Is Fraudulent' column from the DataFrame and stores it in the variable 'col'
# Useful for separating the target variable from feature variables for modeling
```

```python
encoded_mod_df.insert(len(encoded_mod_df.columns),"Is Fraudulent",col)
# Inserts the 'Is Fraudulent' column back into the DataFrame at the last position
# len(encoded_mod_df.columns) ensures it is added at the end
encoded_mod_df
```

```python
encoded_mod_df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
matrix=encoded_mod_df.corr()
# Computes the correlation matrix of the DataFrame
# Shows how strongly each feature is related to the others (values between -1 and 1)

plt.figure(figsize=(30,30))
# Sets the size of the figure to 30x30 inches for better visibility of all features
sns.heatmap(matrix, annot=True, cmap="coolwarm", fmt=".2f")
# annot=True  displays the correlation values on the heatmap
# fmt=".2f"  formats the numbers to 2 decimal places
plt.title("Correlation Heatmap", fontsize=16)
plt.show()
```

```python
Y=encoded_mod_df.iloc[:,-1]
# Selects the last column of the DataFrame as the target variable (Is Fraudulent)

X=encoded_mod_df.iloc[:,0:34]
# Selects the first 34 columns of the DataFrame as feature variables for modeling
```

```python
from sklearn.preprocessing import StandardScaler
# Imports StandardScaler from scikit-learn
# Used to standardize features by removing the mean and scaling to unit variance
to_be_scaled=X.select_dtypes(include=["int64","float64"]).columns.to_list()
# Selects all numeric columns (integers and floats) from X for potential scaling
# Converts the column names to a list
scaling_columns=to_be_scaled[:]
# Creates a copy of the numeric columns list to modify without affecting the original list
for i in to_be_scaled:
  if i in ("Day","Month","Year"):
    scaling_columns.remove(i)
            # Removes date-related columns from scaling, since scaling them may not make sense

  if i in encoded_df.columns:
    scaling_columns.remove(i)
            # Removes one-hot encoded columns from scaling, since they are already 0/1

print(encoded_df.columns)
# Prints the columns of the one-hot encoded DataFrame

print(scaling_columns)
# Prints the final list of columns that will be scaled

print(encoded_mod_df.columns)
# Prints all column names of the full encoded DataFrame
```

```python
scaler=StandardScaler()
# Creates an instance of StandardScaler
# Used to standardize numerical features by removing the mean and scaling to unit variance
X[scaling_columns]=scaler.fit_transform(X[scaling_columns])
# Applies the scaler to the selected numeric columns in X
# fit_transform learns the mean and standard deviation, then scales the data
X
```

Spliting datasets

```python
from sklearn.model_selection import train_test_split
# Imports the train_test_split function from scikit-learn
x_train,x_test,y_train,y_test=train_test_split(X,Y,stratify=Y,test_size=0.3,random_state=52)
# Ensures the train and test sets have the same proportion of target classes
# Splits the feature matrix X and target vector Y into training and testing sets
```

```python
from sklearn.linear_model import LogisticRegression
# Used to build a logistic regression model for classification tasks
```

```python
lor = LogisticRegression(
    max_iter=1000,         # Maximum number of iterations for the solver to converge
    penalty="elasticnet",  # Uses Elastic Net regularization (combination of L1 and L2)
    solver="saga",         # Solver that supports Elastic Net penalty
    l1_ratio=0.5,          # Balance between L1 (lasso) and L2 (ridge) regularization (0.5 = equal mix)
    class_weight="balanced" # Automatically adjusts weights inversely proportional to class frequencies
)
# Creates a Logistic Regression model instance with specified hyperparameters
```

```python
lor.fit(x_train,y_train)
# Trains (fits) the Logistic Regression model on the training data
# x_train  feature variables for training
# y_train  target variable for training
```

```python
y_pred=lor.predict(x_test)
```

```python
train_acc=lor.score(x_train,y_train)
# Calculates the accuracy of the Logistic Regression model on the training set
# Compares predicted values with actual values in y_train

print(train_acc)
```

Adding Smote to x and y

```python
from imblearn.over_sampling import SMOTE
# Imports SMOTE (Synthetic Minority Over-sampling Technique) from imbalanced-learn
# Used to handle imbalanced datasets by generating synthetic samples for the minority class
sm=SMOTE()
# Creates an instance of SMOTE with default parameters
# Will be used to balance the target classes in the dataset
```

```python
x_sm_train,y_sm_train=sm.fit_resample(x_train,y_train)
# Applies SMOTE to the training data to handle class imbalance
# fit_resample learns the minority class distribution and generates synthetic samples
# x_sm_train  new balanced feature set for training
# y_sm_train  new balanced target set for training
```

Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(max_iter=1000,penalty="l2")
# Creates a new Logistic Regression model
# max_iter=1000  allows up to 1000 iterations for convergence
# penalty="l2"  applies L2 (Ridge) regularization to prevent overfitting
lr.fit(x_sm_train,y_sm_train)
# Trains the Logistic Regression model on the balanced training data
# x_sm_train  features after SMOTE
# y_sm_train  target after SMOTE
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, classification_report, confusion_matrix
# Imports evaluation metrics from scikit-learn to assess model performance
# accuracy_score  measures overall correctness of predictions
# precision_score  measures how many predicted positives are actually positive
# classification_report  provides precision, recall, f1-score, and support for each class
# confusion_matrix  shows a matrix of true vs predicted labels for detailed error analysis
```

```python
print(classification_report(y_test,y_pred))
# Prints a detailed classification report comparing actual vs predicted values
# Includes metrics for each class: precision, recall, f1-score, and support
# Helps evaluate the performance of the model on the test set
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(precision_score(y_test,y_pred))
```

```python
df=pd.read_csv("/content/data_FinancialTransactions.csv")
df.head()
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df.head()
```

Date conversion

```python
df["Date"]=pd.to_datetime(df["Date"])
# Converts the 'Date' column in the DataFrame 'df' from string/object to datetime format
# Enables time-based operations like extracting day, month, year, or filtering by date
```

```python
day_loc=df.columns.get_loc("Date")+1
day_values=df["Date"].dt.day
df.insert(day_loc,"Day",day_values)
df
```

```python
month_loc=df.columns.get_loc("Day")+1
month_values=df["Date"].dt.month
df.insert(month_loc,"Month",month_values)
df
```

```python
year_loc=df.columns.get_loc("Month")+1
year_values=df["Date"].dt.year
df.insert(year_loc,"Year",year_values)
df
```

```python
df.drop("Date",axis=1,inplace=True)
df.head()
```

Encoding

Encoding for ordinal data

```python
frequency_map={"Low":0,"Medium":1,"High":2}
Reputation_map={"Bad":0,"Average":1,"Good":2}
df["Online Transactions Frequency"]=df["Online Transactions Frequency"].map(frequency_map)
df["Merchant Reputation"]=df["Merchant Reputation"].map(Reputation_map)
df
```

Encoding for nominal data

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
# sparse_output=False ensures the output is a dense NumPy array instead of a sparse matrix
```

```python
nominal_cols=df.select_dtypes(include=["object"]).columns.tolist()
encoded=ohe.fit_transform(df[nominal_cols])
encoded_df=pd.DataFrame(encoded,columns=ohe.get_feature_names_out(nominal_cols))
encodedmod_df=pd.concat([df.drop(columns=nominal_cols),encoded_df],axis=1)
encodedmod_df.columns
```

```python
col=encodedmod_df.pop("Is Fraudulent")
```

```python
encodedmod_df.insert(len(encodedmod_df.columns),"Is Fraudulent",col)
# Inserts the 'Is Fraudulent' column at the end of the DataFrame
# len(encodedmod_df.columns) ensures it is added as the last column
# 'col' contains the original target variable separated earlier
encodedmod_df
```

```python
fig=plt.figure(figsize=(30,30))
mat=encodedmod_df.corr()
# Shows how strongly each feature is related to the others (values between -1 and 1)

sns.heatmap(mat,annot=True)
plt.tight_layout()
# Adjusts subplot parameters to give the plot proper spacing and prevent overlap
```

```python
sns.countplot(x="Year",data=encodedmod_df)
# Creates a count plot using seaborn to show the number of transactions per year
# x="Year"  plots counts for each unique year
```

```python
encodedmod_df.drop("Year",axis=1,inplace=True)
```

```python
X=encodedmod_df.iloc[:,0:len(encodedmod_df)-1]
# Selects all columns except the last one as feature variables (inputs) for modeling

Y=encodedmod_df.iloc[:,-1]
# Selects the last column ('Is Fraudulent') as the target variable (output) for modeling
```

```python
from sklearn.model_selection import train_test_split
```

```python
tree_x_train,tree_x_test,tree_y_train,tree_y_test=train_test_split(X,Y,stratify=Y,random_state=52,test_size=0.4)
```

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(
    class_weight="balanced",    # Adjusts weights inversely proportional to class frequencies to handle imbalance
    criterion="entropy",        # Uses information gain (entropy) to measure the quality of a split
    random_state=42,            # Sets a seed for reproducibility
    max_depth=5,                # Limits the maximum depth of the tree to prevent overfitting
    min_samples_split=10,       # Requires at least 10 samples to split a node
    max_features='sqrt'         # Considers the square root of the number of features at each split
)
# Creates an instance of a decision tree classifier with specified hyperparameters
```

```python
tree.fit(tree_x_train,tree_y_train)
```

```python
y_pred=tree.predict(tree_x_test)
```

```python
print(classification_report(y_pred,tree_y_test))
```

```python
print(confusion_matrix(y_pred,tree_y_test))
```

```python
print(accuracy_score(y_pred,tree_y_test))
```

```python
from imblearn.over_sampling import SMOTE
# Imports SMOTE (Synthetic Minority Over-sampling Technique) from the imbalanced-learn library
# Used to balance imbalanced datasets by generating synthetic samples for the minority class
```

```python
sm=SMOTE()
X_sampled,Y_sampled=sm.fit_resample(tree_x_train,tree_y_train)
# Applies SMOTE to the training data
# Generates synthetic samples for the minority class to balance the dataset
# X_sampled  new feature set after oversampling
# Y_sampled  new target set after oversampling
```

```python
sample_tree=DecisionTreeClassifier(class_weight="balanced",random_state=42,max_depth=10,max_features="sqrt")
# Considers the square root of total features at each split
```

```python
sample_tree.fit(X_sampled,Y_sampled)
```

```python
sampled_y_pred=sample_tree.predict(tree_x_test)
```

```python
print(classification_report(sampled_y_pred,tree_y_test))
```

```python
print(accuracy_score(sampled_y_pred,tree_y_test))
```

```python
print(confusion_matrix(sampled_y_pred,tree_y_test))
```
